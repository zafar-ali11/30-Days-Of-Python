**What Is Indexing in Strings?**

Indexing means giving each character a position number, so Python knows where to find it.

```python
gene_sequence = "ATGCGTTAA"
print("first ", gene_sequence[0])
print("second ", gene_sequence[1])
print("third ", gene_sequence[2])
#extract start and stop codon from the sequence
print("startcodon, [0:3]")
print("stopcodon, [-3]")
if gene_sequence.startswith("ATG") and gene_sequence.endswith("TAA"):
    print("this is a valid gene sequence")
else:
    print("this is not a valid sequence")
```

Identifying Mutations

```python
reference = "ATGCGTTAA"
mutated   = "ATGCATTAA"  # base mutation at index 4

if reference[4] != mutated[4]:
    print("Mutation detected at position 4:", reference[4], "→", mutated[4])
```

Count the number of nucleotides in a sequence

```python
dna = "ATGCTTAGGCAT"
a_count=dna.count("A")
t_count= dna.count("T")
g_count= dna.count("G")
print("A:",a_count)
print("T:",t_count)
print("G:",g_count)
```
